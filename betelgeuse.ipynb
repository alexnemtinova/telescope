# SkyView

```python
import astropy.units as u
from astroquery.skyview import SkyView
skyv = SkyView()
wav_nam = ['WISE 3.4', 'WISE 4.6', 'WISE 12', 'WISE 22']
hduls = skyv.get_images(position='Betelgeuse', radius=5*u.arcmin,
                        survey=wav_nam)
```

```python
import matplotlib.pyplot as plt
from astropy.wcs import WCS
from astropy.visualization import astropy_mpl_style

plt.style.use(astropy_mpl_style)

fig = plt.figure(figsize=(10, 10))
for i in range(4):
    if i<2: ax = plt.subplot2grid((2,2),(0,i),projection=WCS(hduls[i][0].header))
    else: ax = plt.subplot2grid((2,2),(1,i-2),projection=WCS(hduls[i][0].header)) 
    ax.imshow(hduls[i][0].data)
    ax.set_title(wav_nam[i])
    plt.xlabel('RA')
    plt.ylabel('Dec')
plt.subplots_adjust(hspace=0.4)

plt.show()
```
